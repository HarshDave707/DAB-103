```python
import pandas as pd

vehicle_dataset_url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv'

vehicle_df = pd.read_csv(vehicle_dataset_url)

vehicle_df
```

```python
vehicle_df.drop ( columns= ['VIN (1-10)',
                            'Base MSRP',
                            'Legislative District',
                            'Vehicle Location',
                            'Vehicle Location',
                            'Electric Utility',
                            '2020 Census Tract'],inplace=True)
```

```python
vehicle_df
```

```python
null_counts = vehicle_df.isnull().sum()
print(null_counts)
```

```python
vehicle_df['County'].fillna(value="not defined", inplace=True)
vehicle_df['City'].fillna(value="not defined", inplace=True)
vehicle_df['Postal Code'].fillna(value="not defined", inplace=True)
vehicle_df['Model'].fillna(value="not defined", inplace=True)

```

```python
null_counts = vehicle_df.isnull().sum()
print(null_counts)
```

```python
charging_station_dataset_url = ('https://developer.nrel.gov/api/alt-fuel-stations/v1.csv?'
       'api_key=5gylZXuJK3qHQRwp8SyfTgtlomhXSCWSlfyA7y3P'
       '&status=E'
       '&access=public'
       '&fuel_type=ELEC'
       '&country=CA'
       )

charging_station_vehicle_df = pd.read_csv(charging_station_dataset_url)

charging_station_vehicle_df
```

```python
charging_station_vehicle_df.drop(columns=['Intersection Directions', 'Plus4', 'Expected Date', 
                                          'Groups With Access Code', 'Cards Accepted', 'BD Blends', 
                                          'NG Fill Type Code', 'NG PSI', 'EV Other Info', 
                                          'EV Network', 'EV Network Web', 'Date Last Confirmed', 
                                          'Updated At', 'Federal Agency ID', 'Federal Agency Name', 
                                          'Open Date', 'Hydrogen Status Link', 'NG Vehicle Class', 
                                          'LPG Primary', 'E85 Blender Pump', 'Intersection Directions (French)', 
                                          'Access Days Time (French)', 'BD Blends (French)', 'Groups With Access Code (French)', 
                                          'Hydrogen Is Retail', 'Access Code', 'Access Detail Code', 
                                          'Federal Agency Code', 'Facility Type', 'CNG Dispenser Num', 
                                          'CNG On-Site Renewable Source', 'CNG Total Compression Capacity', 
                                          'CNG Storage Capacity', 'LNG On-Site Renewable Source', 
                                          'E85 Other Ethanol Blends', 'EV Pricing', 'EV Pricing (French)', 
                                          'LPG Nozzle Types', 'Hydrogen Pressures', 'Hydrogen Standards', 
                                          'CNG Fill Type Code', 'CNG PSI', 'CNG Vehicle Class', 'LNG Vehicle Class', 
                                          'EV On-Site Renewable Source', 'Restricted Access', 'RD Blends', 
                                          'RD Blends (French)', 'RD Blended with Biodiesel', 'RD Maximum Biodiesel Level', 
                                          'NPS Unit Name'], inplace=True)
```

```python
charging_station_vehicle_df
```

```python
null_counts = charging_station_vehicle_df.isnull().sum()
print(null_counts)
```

```python
charging_station_vehicle_df['State'].fillna(value="not defined", inplace=True)
```

```python
#Station Phone
charging_station_vehicle_df['Station Phone'] = charging_station_vehicle_df['Station Phone'].fillna(method='bfill')
```

```python
#Access Days Time
charging_station_vehicle_df['Access Days Time'] = charging_station_vehicle_df['Access Days Time'].fillna(method='ffill')
```

```python
#EV Level2 EVSE Num
mean_value = charging_station_vehicle_df['EV Level2 EVSE Num'].mean()
charging_station_vehicle_df['EV Level2 EVSE Num'].fillna(mean_value, inplace=True)
```

```python
#Geocode Status
charging_station_vehicle_df['Geocode Status'] = charging_station_vehicle_df['Geocode Status'].fillna(method='ffill')
```

```python
#Owner Type Code,EV DC Fast Count ,EV Level1 EVSE Num
charging_station_vehicle_df['Owner Type Code'].fillna(value="not defined", inplace=True)
charging_station_vehicle_df['EV DC Fast Count'].fillna(value="not defined", inplace=True)
charging_station_vehicle_df['EV Level1 EVSE Num'].fillna(value="not defined", inplace=True)
```
