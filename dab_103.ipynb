```python
# trying  to create
import folium

# Create a map centered on Washington state
wa_map = folium.Map(location=[47.7511, -120.7401], zoom_start=7)

# Define a dictionary that maps fuel types to colors
fuel_colors = {'ELEC': 'green', 'LPG': 'red', 'Other': 'blue'}

# Add a marker for each charging station
for index, row in wa_fuel_df.iterrows():
    # Get the coordinates and fuel type for the station
    lat = row['Latitude']
    lon = row['Longitude']
    fuel_type = row['Fuel Type Code']
    
    # Determine the color of the marker based on the fuel type
    color = fuel_colors.get(fuel_type, 'gray')
    
    # Add the marker to the map
    folium.Marker(location=[lat, lon], icon=folium.Icon(color=color)).add_to(wa_map)

# Display the map
wa_map
```
