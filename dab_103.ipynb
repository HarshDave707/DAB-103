```python
import pandas as pd

vehicle_dataset_url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv'

vehicle_df = pd.read_csv(vehicle_dataset_url)

vehicle_df
```

```python
vehicle_df.drop ( columns= ['VIN (1-10)',
                            'Base MSRP',
                            'Legislative District',
                            'Vehicle Location',
                            'Vehicle Location',
                            'Electric Utility',
                            '2020 Census Tract'],inplace=True)
```

```python
vehicle_df
```

```python
null_counts = vehicle_df.isnull().sum()
print(null_counts)
```

```python
null_County_row = vehicle_df[vehicle_df['County'].isnull()]
vehicle_df = vehicle_df.dropna(subset=['County'])
```

```python

vehicle_df['Model'].fillna(value="not defined", inplace=True)

```

```python
null_counts = vehicle_df.isnull().sum()
print(null_counts)
```
